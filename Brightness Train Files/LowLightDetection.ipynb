```python
import numpy as np
import cv2

def low_light_detection(img):
    img = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
    x = np.mean(img[:, :, 2])
    return x
```

```python
import os
```

```python
high = [cv2.imread('./our485/high/'+i) for i in sorted(os.listdir('./our485/high/')) if '.DS_Store' not in i]
low = [cv2.imread('./our485/low/'+i) for i in sorted(os.listdir('./our485/low/')) if '.DS_Store' not in i]

print(len(high))
print(len(low))
x = [low_light_detection(i) for i in high]+[low_light_detection(i) for i in low]
y = [0 for i in high]+[1 for i in low]

x, y = np.array(x), np.array(y)
indexes = np.arange(len(x))
np.random.shuffle(indexes)
x = x[indexes]
y = y[indexes]
```

```python
x = x.reshape((-1, 1))
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
clf = LogisticRegression(random_state=0).fit(x, y)
```

```python
clf.score(x, y)
```

```python
np.hstack((clf.intercept_[:,None], clf.coef_))
```

```python
def low_light_detection(img):
    img = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
    x = np.mean(img[:, :, 2])
    z = x*(-0.14439479)+9.92969547
    return 1/(1+np.exp(-z))
```
